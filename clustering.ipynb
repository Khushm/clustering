```python
# Importing necessary libraries for data manipulation, clustering, evaluation, and visualization

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score, adjusted_rand_score, homogeneity_score, classification_report, confusion_matrix
from scipy.optimize import linear_sum_assignment
import scipy.cluster.hierarchy as sch
import time
from sklearn.preprocessing import StandardScaler
```

This function uses the Hungarian algorithm (through linear_sum_assignment) to optimally match predicted labels to true labels by maximizing the number of matched points. It builds a cost matrix where each cell represents how many points are correctly classified between a true and predicted label. The function then returns the matched predicted labels based on the optimal assignment of clusters.

```python
# Function to match predicted and true labels using Hungarian algorithm
def match_labels(true_labels, predicted_labels, n_clusters):
    # Create a cost matrix with size (n_clusters x n_clusters)
    cost_matrix = np.zeros((n_clusters, n_clusters))
    
    # Populate the cost matrix with the number of matching points between true and predicted labels
    for i in range(n_clusters):
        for j in range(n_clusters):
            # Count the number of points that belong to both cluster 'i' (true) and cluster 'j' (predicted)
            cost_matrix[i, j] = np.sum((true_labels == i) & (predicted_labels == j))
    
    # Use the Hungarian algorithm (linear sum assignment) to find the optimal label matching
    row_ind, col_ind = linear_sum_assignment(cost_matrix, maximize=True)
    
    # Apply the optimal label matching to the predicted labels
    matched_labels = np.array([col_ind[label] for label in predicted_labels])
    
    return matched_labels
```

It computes the Within-cluster Sum of Squares (WSS), Between-cluster Sum of Squares (BSS), and Total SSE to evaluate clustering performance. WSS measures how tightly data points are grouped within each cluster, BSS quantifies the separation between clusters, and Total SSE is the sum of WSS and BSS, representing overall clustering quality.

```python
def calculate_true_clusters(X, true_labels):
    # Get the unique labels of the clusters
    unique_labels = np.unique(true_labels)
    
    # Get the total number of data points
    n = X.shape[0]
    
    # Calculate the overall mean of the dataset
    total_mean = np.mean(X, axis=0)
    
    # Initialize WSS and BSS to zero
    WSS = 0
    BSS = 0
    
    # Loop through each cluster label
    for label in unique_labels:
        # Get the points that belong to the current cluster
        cluster_points = X[true_labels == label]
        
        # Calculate the mean of the current cluster
        cluster_mean = np.mean(cluster_points, axis=0)
        
        # Compute WSS: Sum of squared distances between each point and the cluster's mean
        WSS += np.sum(np.linalg.norm(cluster_points - cluster_mean, axis=1) ** 2)
        
        # Compute BSS: Sum of squared distances between the cluster's mean and the overall mean,
        # weighted by the number of points in the cluster
        BSS += len(cluster_points) * np.linalg.norm(cluster_mean - total_mean) ** 2
    
    # Total SSE is the sum of WSS and BSS
    total_SSE = WSS + BSS
    return WSS, BSS, total_SSE
```

I have used this same kmeans_clustering function for each run of the K-means clustering with different values of K (the number of clusters). For each configuration of K, the function calculates the WSS, BSS, and total SSE values, and also tracks the execution time of the K-means algorithm. The function is reused across multiple values of K to maintain consistency and streamline the process of clustering evaluation.


**Paramaters** - <br>
n_clusters=K: Sets the number of clusters. Chose K to match the number of true classes in your data.<br>
init='k-means++': Picks better initial centroids, leading to faster and more accurate results. (default)<br>
max_iter=300: Limits the number of iterations to 300, enough for most cases to ensure convergence. (default)<br>
random_state=42: Ensures results are the same each run, making comparisons consistent and reproducible.

```python
# Function to perform K-means clustering and calculate WSS, BSS, and total SSE
def kmeans_clustering(X, K):   
    # Start the timer to calculate execution time
    start_time = time.time()
    
    # Initialize KMeans with specified parameters and fit the model to the data
    kmeans = KMeans(n_clusters=K, init='k-means++', max_iter=300, random_state=42)
    kmeans.fit(X)
    
    # Get the predicted labels after fitting the model
    predicted_labels = kmeans.labels_
    
    # Calculate WSS, BSS, and total SSE using the true clustering method
    WSS, BSS, total_SSE = calculate_true_clusters(X, predicted_labels)
    
    # End the timer and calculate the total execution time
    end_time = time.time()
    execution_time = end_time - start_time
    
    # Return the calculated metrics and execution time
    return WSS, BSS, total_SSE, execution_time
```

```python
# Load the dataset
data = pd.read_csv('small_Xydf.csv')
X = data[['X0', 'X1']].values
true_labels = data['y'].values

WSS, BSS, total_SSE = calculate_true_clusters(X, true_labels)
print(f"True WSS: {round(WSS, 4)}\nTrue BSS: {round(BSS, 4)}\nTotal SSE: {round(total_SSE, 4)}")
```

```python
# Run K-means for K=2, 3, 4
results = {}
for K in [2, 3, 4]:
    WSS, BSS, total_SSE, execution_time = kmeans_clustering(X, K)
    results[K] = {'WSS': round(WSS, 4), 'BSS': round(BSS, 4), 'Total SSE': round(total_SSE, 4), 'Execution Time': round(execution_time, 4)}

for key, value in results.items():
    print("For K value =", key, ":", value)
```

```python
# Function to perform DBSCAN and calculate WSS, BSS
def dbscan_clustering(X, eps, min_samples):
    start_time = time.time()
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    dbscan.fit(X)
    
    # Assign the predicted labels
    predicted_labels = dbscan.labels_
    
    # If there are any noise points (-1), exclude them for evaluation
#     if np.any(predicted_labels == -1):
#         print("DBSCAN identified some noise points as -1. These will be excluded.")
    
    # Calculate WSS and BSS (need to define a function to calculate these)
    WSS, BSS, total_SSE = calculate_true_clusters(X, predicted_labels)
    
    end_time = time.time()
    execution_time = end_time - start_time
    
    return WSS, BSS, total_SSE, execution_time, predicted_labels
```

I started with **eps = 0.1, min_samples = 20** because it balanced tight clusters and avoided noise, resulting in clear, well-defined clusters. With **eps = 0.3, min_samples = 10**, the larger `eps` overestimated cluster boundaries, adding incorrect groupings. Finally, **eps = 0.2, min_samples = 2** was tested but resulted in noise due to the low `min_samples`. Overall, **eps = 0.1, min_samples = 20** gave the best results, offering tight clusters without excess noise.

```python
# Function to perform Agglomerative Clustering and plot the dendrogram
def dendrogram_clustering(X, n_clusters):
    start_time = time.time()
    
    # Fit Agglomerative Clustering
    agglomerative = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
    predicted_labels = agglomerative.fit_predict(X)
    
    # Compute WSS, BSS, and total SSE (using the same function from previous)
    WSS, BSS, total_SSE = calculate_true_clusters(X, predicted_labels)
    
    end_time = time.time()
    execution_time = end_time - start_time
    
    return WSS, BSS, total_SSE, execution_time, predicted_labels
```
